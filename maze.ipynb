```python
#prim's algorithm - maze generation
import random
points = list(range(1,101))
graph = {}
for point in points:
    graph[point] = []
    if point % 10 != 0:
        graph[point].append((point+1,random.randint(1,100)))
    if point % 10 != 1:
        graph[point].append((point-1,random.randint(1,100)))
    if point > 10:
        graph[point].append((point-10,random.randint(1,100)))
    if point < 91:
        graph[point].append((point+10,random.randint(1,100)))

def prims_fast(G):
    from heapq import heapify, heappop, heappush
    s = list(G.keys())[0]
    E = {s}
    HEAP = []
    cost = []
    for i in G[s]:
        heappush(HEAP,(i[1],s,i[0]))
    while len(E) != len(G):
        length,previous_node,node = heappop(HEAP)
        if node in E:
            continue
        E.add(node)
        cost.append((node,previous_node,length))
        for i,length in G[node]:
            if i not in E:
                heappush(HEAP,(length,node,i))
    return cost

lines = prims_fast(graph)

coordinates = {}
for point in points:
    coordinates[point] = (50+50*((point-1)%10),50+50*((point-1)//10))

import pygame
from pygame.locals import *
pygame.init()
window = pygame.display.set_mode((550, 550))
window.fill((0, 0, 0))

for a,b,i in lines:
    pygame.draw.line(window, (255, 255, 255), [coordinates[a][0], coordinates[a][1]], [coordinates[b][0], coordinates[b][1]],25)

for i in coordinates:
    pygame.draw.rect(window,(255,255,255),[coordinates[i][0]-12,coordinates[i][1]-12,25,25])

pygame.draw.rect(window,(0,0,255),[coordinates[1][0]-12,coordinates[1][1]-12,25,25])
pygame.draw.rect(window,(255,0,0),[coordinates[100][0]-12,coordinates[100][1]-12,25,25])
    
pygame.display.update()

```

```python
#DFS - maze solving
new = {}
for i in range(1,101):
    new[i] = []
for a,b,c in lines:
    new[a].append(b)
    new[b].append(a)

def DFS_iter(G, s):
    paths = {s:[s]}
    E = set()
    stack = [s]
    while stack:
        x = stack.pop()
        for node in G[x]:
            if node not in E:
                stack.append(node)
                E.add(node)
                paths[node] = paths[x] + [x,node]
    

    
    return paths[100]

path = DFS_iter(new,1)

for i in range(len(path)-1):
    pygame.draw.line(window, (255, 255, 0), [coordinates[path[i]][0], coordinates[path[i]][1]], [coordinates[path[i+1]][0], coordinates[path[i+1]][1]],25)
    pygame.draw.rect(window,(255,255,0),[coordinates[path[i]][0]-12,coordinates[path[i]][1]-12,25,25])

pygame.draw.rect(window,(0,0,255),[coordinates[1][0]-12,coordinates[1][1]-12,25,25])
pygame.draw.rect(window,(255,0,0),[coordinates[100][0]-12,coordinates[100][1]-12,25,25])
    
pygame.display.update()
```
